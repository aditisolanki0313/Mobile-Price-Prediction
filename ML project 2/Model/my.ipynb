```python
df['Battery_mAh'] = df['Battery'].str.extract(r'(\d+)').astype(int)
df.head()
```

```python
df1 = df.drop(['Battery'], axis='columns')
df1.head(10)
```

```python
df1['Camera_MP'] = df1['Camera'].str.extract(r'(\d+)').astype(int)
df1['Storage_GB_RAM'] = df1['Storage'].str.extract(r'(\d+)').astype(int)
df1.head()
```

```python
df1['Storage_GB_ROM'] = df1['Storage'].str.extract(r'\d+\D+(\d+)').astype('Int64')
df1.head()
```

```python
df1.Processor.isnull().sum()
```

```python
df2 = df1.drop('Processor', axis='columns')
```

```python
df2['Display_cm'] = df2['Display'].str.extract(r'(\d+)').astype(int)
df2.head()
```

```python
df3 = df2.drop(['Storage', 'Camera', 'Display','ImageURL'], axis='columns')
df3.head()
```

MACHINE LEARNING

```python
Name_list = df3['Name'].unique().tolist()
Name_list
```

```python
Name_list_encoded = pd.get_dummies(Name_list, columns=['Name'])
Name_list_encoded = Name_list_encoded.astype(int)
Name_list_encoded.head()
```

```python
df4 = pd.concat([df3, Name_list_encoded], axis='columns')
df4.head()
```

```python
df5 = df4.dropna()
df5.isnull().sum()
```

```python
df6 = df5.drop('Name', axis='columns')
df6.head()
```

```python
df6['Price(Rs)'] = df6['Price(Rs)'].str.replace(',', '')
df6['Price(Rs)'] = pd.to_numeric(df6['Price(Rs)'])
df6.head()
```

```python
df6.rename(columns={'Price(Rs)': 'Price'}, inplace=True)
df6.head()
```

```python
df6.shape
```

```python
x = df6.drop('Price', axis='columns')
x
```

```python
y = df6['Price']
y
```

```python
from sklearn.model_selection import train_test_split 
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression 
lr_clf = LinearRegression()
lr_clf.fit(x_train,y_train)
lr_clf.score(x_test,y_test) 
```

```python
from sklearn.model_selection import ShuffleSplit 
from sklearn.model_selection import cross_val_score 
cv = ShuffleSplit(n_splits=4, test_size= 0.2, random_state=0)

cross_val_score(LinearRegression(), x,y, cv=cv)
```

```python
from sklearn.model_selection import GridSearchCV 
from sklearn.linear_model import Lasso 
from sklearn.tree import DecisionTreeClassifier 

def best_model_gridsearchCV(x,y):
    algos = {
        'linear_regression': {
            'model': LinearRegression(),
            'params' :{
                     'positive': [True,False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree':{
            'model': DecisionTreeClassifier(),
            'params': {
                'criterion': ['mse', 'friedman_mse'],
                'spitter': ['best', 'random']
            }
        }
    }


    scores= []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for algo_name, config in algos.items():
        gs = GridSearchCV(config['model'],config['params'], cv=cv, return_train_score=False)
        gs.fit(x,y)
        scores.append({
            'model': algo_name,
            'best_score': gs.best_score_,
            'best_params': gs.best_params_
        })

        return pd.DataFrame(scores,columns = ['model','best_score','best_params'])
    
    print(best_model_gridsearchCV(x,y))
```

```python
def predict_Price(Name, Battery_mAh, Camera_MP, Storage_GB_RAM,Storage_GB_ROM,Display_cm):
    loc_index = np.where(x.columns==Name)[0][0]

    x_input = np.zeros(len(x.columns))

    x_input[0] = Battery_mAh
    x_input[1] = Camera_MP
    x_input[2] = Storage_GB_RAM
    x_input[3] = Storage_GB_ROM
    x_input[4] = Display_cm
    # x_input[5] = Processor_mob

    if loc_index >= 0:
        x_input[loc_index] = 1

    return lr_clf.predict([x_input])[0]


    
```
